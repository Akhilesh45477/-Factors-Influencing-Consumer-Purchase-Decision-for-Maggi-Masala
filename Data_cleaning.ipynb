```python
df=df.drop(["How often do you consume Maggi masala ?","On a scale from 1 (not at all likely) to 10 (extremely likely), how likely are you to buy Maggi Masala again and recommend it to others?"],axis=1)
```

```python
#final dataframe for factor analysis
type(df)
```

```python
#Convert all responses to lowercase
df = df.map(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
#Mapping a dictionary
mapping = {
    "strongly disagree": 1,
    "disagree": 2,
    "neutral": 3,
    "agree": 4,
    "strongly agree": 5
}
```

```python
#Replacing Values using mappings
df = df.map(lambda x: mapping.get(x, x))
```
