```python
import pandas as pd
import numpy as np
from factor_analyzer import FactorAnalyzer
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("cleaned_responses.csv")
```

```python
data.columns
```

```python
data.head()
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
```

```python
import statsmodels.api as sm
```

```python
data.dropna(inplace=True)
```

```python
df=data
```

```python
df= (df - df.mean()) / df.std()
df.tail()
```

```python
df.info()
```

```python
#Adequacy Test
#Bartlett’s test of sphericity

from factor_analyzer.factor_analyzer import calculate_bartlett_sphericity
chi_square_value, p_value=calculate_bartlett_sphericity(df)
print("Chi Square Value=",chi_square_value, "P Values=",p_value)
```

```python
# Kaiser-Meyer-Olkin (KMO) Test

from factor_analyzer.factor_analyzer import calculate_kmo
kmo_all,kmo_model=calculate_kmo(df)
print("KMO Value=",kmo_model)
```

```python
# Choosing the Number of Factors

from factor_analyzer import FactorAnalyzer
fa = FactorAnalyzer(rotation = None)
fa.fit(df)
ev,v =fa.get_eigenvalues()
ev
```

```python
# Factor Analysis execution

fa = FactorAnalyzer(n_factors=6,rotation=None, method='principal')
fa.fit(df)

#print(pd.DataFrame(fa.loadings_,index=dataframe.columns))
```

```python
# Factor Analysis execution

fa = FactorAnalyzer(n_factors=4,rotation='varimax', method='principal')
fa.fit(df)

#print(pd.DataFrame(fa.loadings_,index=dataframe.columns))
#with 4 factors(latent Variables)
```

```python
# Factor Loadings

fa.loadings_

floading=pd.DataFrame(fa.loadings_)
floading
```

```python
fa.rotation_matrix_
froma=pd.DataFrame(fa.rotation_matrix_)
froma
```

```python
# Factor Scores

fscore = fa.transform(df)

fscore_data=pd.DataFrame(fscore)
fscore_data
```
